```python
# importing the january data to dataframe using pandas 
import pandas as pd

df_jan = pd.read_csv('https://1.climate.daknob.net/ftp.ncdc.noaa.gov/pub/data/hourly_precip-3240/by_month2013/3240jan2013.dat',header=None, names = ["col1"])

df_jan.shape
```

```python
# FIltering records containing R that as suspect values
df_jan2 =  df_jan[~df_jan['col1'].str.contains('R')]
df_jan2.shape
```

```python
#Printing the head of data
df_jan2.head()
```

```python
#Creating columns by spliting the column 
df_jan2["State"] =df_jan2["col1"].str.slice(3, 5)
df_jan2["year"] =df_jan2["col1"].str.slice(17, 21)
df_jan2["Month"] =df_jan2["col1"].str.slice(21, 23)
df_jan2["Date"] =df_jan2["col1"].str.slice(23, 27)
df_jan2["Units"] = df_jan2["col1"].str.slice(15, 17)
df_jan2["counts"] = df_jan2["col1"].str.slice(27, 30)
# Extracting the last time and value for each row  as they represent the final value of the day
df_jan2['time'] = [x.split()[-2] for x in df_jan2.col1]
df_jan2['val'] = [x.split()[-1] for x in df_jan2.col1]
```

```python
#printing the head of dataframe
df_jan2.head()
```
